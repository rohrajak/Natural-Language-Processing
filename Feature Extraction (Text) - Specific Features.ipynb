```python
import nltk
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
from nltk.corpus import stopwords
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.datasets import  fetch_20newsgroups
import re
import string
import pandas as pd
```

```python
newsgroup_data_sample = fetch_20newsgroups(subset='train')
```

Storing the Corpus in DataFrame

```python
lemmatizer = WordNetLemmatizer()
newsgroup_text_df = pd.DataFrame({'text': newsgroup_data_sample['data']})
newsgroup_text_df.head()
```

Clean the data

```python
nltk.download('wordnet')
```

```python
stop_words = stopwords.words('english')
stop_words = stop_words + list(string.printable)
newsgroup_text_df['cleaned_text'] = newsgroup_text_df['text'].apply(\
lambda x: ' '.join([lemmatizer.lemmatize(word.lower())\
for word in word_tokenize(re.sub(r'([^\s\w]|_)+',' ',str(x))) if word.lower() not in stop_words]))
```

Create BoW Model

```python
bag_of_words_model = CountVectorizer(max_features= 20)
bag_of_word_df = pd.DataFrame(bag_of_words_model.fit_transform(newsgroup_text_df['cleaned_text']).todense())
bag_of_word_df.columns = sorted(bag_of_words_model.vocabulary_)
bag_of_word_df.head()
```

TF-IDF Model

```python
tfidf_model = TfidfVectorizer(max_features=20)
tfidf_df = pd.DataFrame(tfidf_model.fit_transform(newsgroup_text_df['cleaned_text']).todense())
tfidf_df.columns = sorted(tfidf_model.vocabulary_)
tfidf_df.head()
```

Comparison b/w BoW and Tf-Idf models

```python
#Most Important word as per the BoW model.

rw = 2
list(bag_of_word_df.columns[bag_of_word_df.iloc[rw,:] == bag_of_word_df.iloc[rw,:].max()])
```

```python
#Most Important word as per the TF-IDF model.

rw = 2
list(tfidf_df.columns[tfidf_df.iloc[rw,:] == tfidf_df.iloc[rw,:].max()])
```
