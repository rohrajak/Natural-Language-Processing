```python
import pandas as pd
from string import punctuation
import nltk
nltk.download('tagsets')
from nltk.data import load
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag
from nltk import word_tokenize
from collections import Counter
```

```python
tagdict = load('help/tagsets/upenn_tagset.pickle')
list(tagdict.keys())
```

```python
data = pd.read_csv("/content/data.csv", header=0)
```

```python
data.head()
```

```python
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
pos_di = {}

for pos in list(tagdict.keys()):
  pos_di[pos]=[]
  for doc in data['text']:
    di = Counter([j for i, j in pos_tag(word_tokenize(doc))])
    for pos in list(tagdict.keys()):
      pos_di[pos].append(di[pos])

feature_df = pd.DataFrame(pos_di)
feature_df.head()
```

```python
# ✅ Initialize pos_di dictionary with empty lists
pos_di = {tag: [] for tag in tagdict.keys()}

# ✅ Process each document
for doc in data['text']:
    di = Counter([j for i, j in pos_tag(word_tokenize(doc))])  # Count POS tags

    # ✅ Append POS counts safely using .get(pos, 0)
    for pos in tagdict.keys():
        pos_di[pos].append(di.get(pos, 0))

# ✅ Create DataFrame after processing all documents
feature_df = pd.DataFrame(pos_di)
print(feature_df.head())
```

Punctuation marks in each sentences

```python
feature_df['num_of_unique_punctuations'] = data['text'].apply(lambda x:\
len(set(x).intersection(set(punctuation))))
feature_df['num_of_unique_punctuations'].head()
```

Capitalized words in each sentences

```python
feature_df['number_of_capital_words'] = data['text'].apply(lambda x:\
len([word for word in word_tokenize(str(x)) if word[0].isupper()]))
feature_df['number_of_capital_words'].head()
```

UnCapitalized words in each sentences

```python
feature_df['number_of_small_words'] = data['text'].apply(lambda x:\
len([word for word in word_tokenize(str(x)) if word[0].islower()]))
feature_df['number_of_small_words'].head()
```

Number of letter in DataFrame

```python
feature_df['number_of_alphabets']=data['text'].apply(lambda x :\
len([ch for ch in str(x) if ch.isalpha()]))
feature_df['number_of_alphabets'].head()
```

Number of digits in DataFrame

```python
feature_df['number_of_alphabets']=data['text'].apply(lambda x :\
len([ch for ch in str(x) if ch.isdigit()]))
feature_df['number_of_alphabets'].head()
```

Number of words in DataFrame

```python
feature_df['number_of_words']=data['text'].apply(lambda x:\
len(word_tokenize(str(x))))
feature_df['number_of_words'].head()
```

Number of whitespaces in DataFrame

```python
feature_df['number_of_whitespaces']=data['text'].apply(lambda x:\
len(str(x).split(' '))-1)
feature_df['number_of_whitespaces'].head()
```
